# Classification using Decision Tree Algorithm

### Importing Required Modules 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

### Loading Dataset

```python
iris_df = pd.read_csv('datasets/Iris.csv', index_col='Id')
iris_df.head()
```

```python
X = iris_df.drop('Species', axis=1).values
print('Sample for X: ', X[0])
print('X Shape: ', X.shape)
print('No. of Samples in X', X.shape[0])


y = LabelEncoder().fit_transform(iris_df['Species']) 
print(y)

train_x, test_x, train_y, test_y = train_test_split(X, y, test_size=0.25)
print('No of train samples: ', train_x.shape[0])
print('No of test samples: ', test_x.shape[0])
```

```python
sns.pairplot(data=iris_df)
plt.show()
```

### Creating Model 

```python
model = DecisionTreeClassifier()
model.fit(train_x, train_y)

print('Test Scores: ', model.score(test_x, test_y))
```

### Visualizing Decision Tree Model

```python
features = ['Sepal Length (cm)', 'Sepal Width (cm)', 'Petal Length (cm)', 'Petal Width (cm)']
classes = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

plt.figure(figsize=(4,4), dpi=600)
plot_tree(model, feature_names=features, class_names=classes, filled=True, )
plt.show()
```
